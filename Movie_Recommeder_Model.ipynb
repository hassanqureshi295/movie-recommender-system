```python
# We have to convert this genres in simple list form

import ast

def convert(x):
  L = []
  for i in ast.literal_eval(x):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
# Function for the cast column coz we need only three dictionaries from it

import ast

def convert2(x):
  L = []
  counter = 0
  for i in ast.literal_eval(x):
    if counter != 3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
# In the crew column, we only need one list which have the job as a Director

def fetch_director(x):
  L = []
  for i in ast.literal_eval(x):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.overview[0]
```

```python
# The overview column is string so we have to convert it to a list in order to merge with other columns

movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# We have to remove all the spaces in the column

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# Merging all the columns in a new column name tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# Extracting the required columns
data = movies[['movie_id', 'title', 'tags']]
```

```python
data.head()
```

```python
# Converting the tag column into string

data['tags'] = data['tags'].apply(lambda x:" ".join(x))
```

```python
data.head()
```

```python
data.info()
```

```python
data.tags[0]
```

```python
# Converting the tags column into lowercase

data['tags'] = data['tags'].apply(lambda x : x.lower())
```

```python
data.head()
```

```python
# This library will perform staming(convert the similar words into one standard word)
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# This is a fuction that will use the nltk.stem library to perform stamming.
# For example: words in the tags like loving, loved and lovely will convert into love

def stem(x):
  y = []

  for i in x.split():
    y.append(ps.stem(i))
  return " "  .join(y)
```

```python
stem("In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron")
```

```python
data["tags"] = data['tags'].apply(stem)
```

Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
# Stop Words parameter will remove all the unuse words from the tags like is, are, from, to etc
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
cv.fit_transform(data['tags']).toarray().shape
```

```python
vectors
```

```python
# These are the words that are common in all the tags

cv.vocabulary_
```

```python
# Now, we have to calculate the distance and simlarities between each vector to each other

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
# Similarity of first movie with all other movies. 1 is the more similarity and 0 is far similarity and with their indexes coz we have to hold the indexes
# We only want those values that have the most similarities with the first movies so that we are using the sorted function
# Extracting the first 5 that have more similarities
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
data[data['title'] == 'Batman Begins'].index[0]
```

```python
# making a recommended function
def recommend(movie):
  movie_index = data[data['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
  # Priniting the first movies title
  for i in movies_list:
    print(data.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
